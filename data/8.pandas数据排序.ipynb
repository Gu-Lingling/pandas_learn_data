### 1.读取数据

```python
import pandas as pd
```

```python
df = "C:/Users/THE KEY/Desktop/python_datum/pandas/data/weather/weather_beijing.xlsx"
wea = pd.read_excel(df)
wea = pd.DataFrame(wea)
```

```python
wea.loc[:,'最高温'] = wea['最高温'].str.replace("°","").astype(int)
wea.loc[:,'最低温'] = wea['最低温'].str.strip("°")
```

```python
wea.loc[[0],"最低温"] = '-7'
wea['最低温'][455]
wea.loc[[455],"最低温"] = '0'
wea.loc[:,'最低温'] = wea['最低温'].astype(int)
```

```python
wea.set_index('日期', inplace= True)
wea.head()
```

### ----------------------------------------------------

### 2.Series排序

```python
# 默认升序排列
wea['最高温'].sort_values()
```

```python
# 降序排序
wea['最低温'].sort_values(ascending = False)
```

```python
# 对中文进行排序
wea['风力风向'].sort_values()
```

### -------------------------------------------------

###  2.DataFrame的排序

```python
# 单列排序
wea.sort_values(by = '最高温', ascending = False)
```

```python
wea.sort_values(by = '最低温', ascending = False)
```

```python
# 多列排序，默认升序
wea.sort_values(by=['空气质量指数','最高温'])
```

```python
wea.sort_values(by = ['空气质量指数','最低温'], ascending = False)
```

```python
# 分别指定升序和降序
wea.sort_values(by=['天气','风力风向'], ascending = [True, False])
```
