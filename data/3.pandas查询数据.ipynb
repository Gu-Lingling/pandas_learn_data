```python
df = "C:/Users/THE KEY/Desktop/python_datum/pandas/data/weather/weather_beijing.xlsx"
```

```python
wea = pd.read_excel(df)
```

```python
wea.loc[:,'最高温'] = wea['最高温'].str.replace("°","").astype(int)
```

```python
wea.loc[:,'最低温'] = wea['最低温'].str.strip("°")
```

```python
type(wea['最低温'][0])
```

```python
# 找出特殊群体nan
i = 0
for x in wea['最低温']:
    i += 1
    if x == 'nan':
        print(x)
        print(i)
#是 0 ，8 ，100，它们是nan，已改正
```

```python
# 手动赋值,先搞成str类型
wea['最低温'][0] = '-7'
type(wea['最低温'][0])
wea['最低温'][8] = '-9'
type(wea['最低温'][8])
wea['最低温'][100] = '9'
type(wea['最低温'][100])
```

```python
# 新方法,不会报错
wea.loc[[0],"最低温"] = '-7'
wea.loc[[8],"最低温"] = '-9'
wea.loc[[100],"最低温"] = '9'
```

```python
# 旧方法

j = 0
k = 0
while k < 4006:
    if type(wea['最低温'][j]) == str:
        num_str = wea['最低温'][j]
        num_int = int(num_str)
        wea['最低温'][j] = num_int
        j += 1
        k += 1
    '''elif wea['最低温'][j] == '':
        wea['最低温'][j] = '0'
        num_str = wea['最低温'][j]
        num_int = int(num_str)
        wea['最低温'][j] = num_int
'''
```

```python
# 新方法
l = 0
m = 0
while m < 4006:
    if type(wea['最低温'][l]) == str:
        num_str = wea['最低温'][l]
        num_int = int(num_str)
        wea.loc[[l],"最低温"] = num_int
        l += 1
        m += 1
```
