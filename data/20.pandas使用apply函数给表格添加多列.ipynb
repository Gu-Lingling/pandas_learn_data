```python
import pandas as pd
```

```python
path = "C:/Users/THE KEY/Desktop/python_datum/pandas/data/weather/weather_beijing.xlsx"
wea = pd.read_excel(path)
wea = pd.DataFrame(wea)
wea.loc[:,'最高温'] = wea['最高温'].str.replace("°","").astype(int)
wea.loc[:,'最低温'] = wea['最低温'].str.strip("°")
wea.loc[[0],"最低温"] = '-7'
wea['最低温'][455]
wea.loc[[455],"最低温"] = '0'
wea.loc[:,'最低温'] = wea['最低温'].astype(int)
# wea.set_index('日期', inplace= True)
wea.head()
```

```python
def wea_func(df):
    return df["最高温"] - df["最低温"], (df["最高温"] + df["最低温"]) / 2

wea[["温差","平均值"]] = wea.apply(wea_func, axis = 1, result_type = "expand")
```
