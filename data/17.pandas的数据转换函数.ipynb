pandas的数据转换函数：map、apply、applymap
1. map：只用于series，实现每个值->值的映射
2. apply：用于series实现每个值的处理，用于df实现某个轴的series的处理
3. applymap：只能用于df，用于处理df中的每个元素**(似乎已经被弃用)

```python
import pandas as pd
stocks = pd.read_excel("C:/Users/THE KEY/Desktop/python_datum/pandas/data/myself_data/stock/stock.xlsx")
stocks.head()
```

```python
stocks["公司"].unique()
```

### --------------------------------------------------------

```python
# map有两种映射Series.map(dict) or Series.map(function)
# 现在把公司映射成中文
# 定义一个字典name
name = {"baba":"阿里巴巴","baidu":"百度","jd":"京东"}
```

```python
# 第一种方法
# lower()把大写变成小写
stocks["公司中文名1"] = stocks["公司"].str.lower().map(name)
stocks.head()
```

```python
# 另一种方法，传函数
stocks["公司中文名2"] = stocks["公司"].map(lambda x:name[x.lower()])
stocks.head()
```

### ---------------------------------------------------------

```python
# apply函数，既可以用于Series，也可以用于DataFrame，参数是传入一个函数
stocks["公司中文名3"] = stocks['公司'].apply(
    lambda x : name[x.lower()])
stocks.head()
```

```python
stocks["公司中文名4"] = stocks.apply(
    lambda x : name[x["公司"].lower()],
    axis = 1)
stocks.head()
```

### -------------------------------------------------------

###  applymap函数适用于DataFrame所有值的转换****(已经被弃用)

```python
# 交易量单位为M
stocks.loc[:,"交易量"] = stocks["交易量"].str.replace("M","")
stocks.head()
```

```python
sub_df = stocks[["收盘","开盘","高","低","交易量"]]
```

```python
sub_df.head()
```

#### 注意：applymap()这个方法在新版的pandas中已经被弃用了，用map()就足够了，意味着map()也可以处理整个DataFrame中所有的值

```python
# 这里的x是整个DataFrame中所有的值
sub_df = sub_df.map(lambda x : int(float(x)))
```
