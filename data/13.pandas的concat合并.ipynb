```python
import pandas as pd
# 忽略报警
import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.DataFrame({'A':['a0','a1','a2','a3'],
                   'B':['b0','b1','b2','b3'],
                   'C':['c0','c1','c2','c3'],
                   'D':['d0','d1','d2','d3'],
                   'E':['e0','e1','e2','e3']
                   })
df1
```

```python
df2 = pd.DataFrame({'A':['a4','a5','a6','a7'],
                   'B':['b4','b5','b6','b7'],
                   'C':['c4','c5','c6','c7'],
                   'D':['d4','d5','d6','d7'],
                   'F':['e4','e5','e6','e7']
                   })
df2
```

```python
# 使用默认的concat进行合并  axis = 0, join = outer, ignore_index = False 
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],ignore_index = True)
```

```python
pd.concat([df1,df2], join = 'inner', ignore_index = True)
```

```python
# 按列添加 axis = 1
s1 = pd.Series(list(range(4)),name = "F")
pd.concat([df1,s1],axis = 1)
```

```python
s2 = df1.apply(lambda x:x["A"]+"_gg", axis = 1)
```

```python
s2
```

```python
s2.name = "G"
```

```python
# 添加多列
pd.concat([df1,s1,s2], axis = 1)
```

```python
# 列表可以只有Series
pd.concat([s1,s2], axis = 1)
```

```python
# 可以混合顺序的
pd.concat([s2,df1,s1], axis = 1)
```
