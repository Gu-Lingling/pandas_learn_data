```python
import pandas as pd
```

### 特殊excel的读取、清洗和处理

```python
# 读取excel，且忽略前面的空行
studf = pd.read_excel("C:/Users/THE KEY/Desktop/python_datum/pandas/data/student/student.xlsx",skiprows = 4)
```

```python
studf
```

```python
# 检测空值
studf.isnull()
```

```python
studf['分数'].isnull()
```

```python
studf['科目'].notnull()
```

```python
# 筛选没有空分数的所有行
studf.loc[studf['分数'].notnull(),:]
```

### -----------------------------------------------------------

```python
# 删掉全是空值的列
studf.dropna(axis = 'columns', how = 'all', inplace = True)
```

```python
studf
```

```python
# 删掉全是空值的行
studf.dropna(axis = 'index', how = 'all', inplace = True)
```

```python
studf
```

###  ----------------------------------------

```python
# 将分数列为空的值替换为0
studf.fillna({"分数":0})
```

```python
# 另一种方法
studf.loc[:,"分数"] = studf["分数"].fillna(0)
```

```python
studf
```

```python
# 将姓名的缺失值填充，根据前面的填充,ffill()   (front)
studf.loc[:,"姓名"] = studf["姓名"].ffill()
```

```python
# 根据后面的填充，在这里是不对的。 函数为bfill()   (behind)
# studf.loc[:,"姓名"] = studf["姓名"].bfill()
```

```python
studf
```

### 将清洗好的excel保存
