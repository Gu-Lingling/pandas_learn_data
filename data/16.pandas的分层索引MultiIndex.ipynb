### -----------------------------------------------------------------

**重要知识**,在选择数据时：
- 元组(key1,key2)代表筛选多层索引，其中key1是索引第一级，key2是第二级，例如key1=JD,key2=2023/3/27
- 列表[key1,key2]代表同一层的多个key，其中key1和key2是并列的同级索引，比如key1=JD,key2=BAIDU

```python
stocks.loc["JD"].head(3)
```

```python
stocks.loc[("JD","2023-03-08"),:]
```

```python
stocks.loc[("JD","2023-03-08"),"交易量"]
```

```python
# 并列的
stocks.loc[["BAIDU","JD"],:]
```

```python
stocks.loc[(["JD","BAIDU"],"2023-03-08"),:]
```

```python
stocks.loc[(["JD","BAIDU"],"2023-03-08"),'涨跌幅']
```

```python
stocks.loc[('BABA',["2023-02-28","2023-03-27"]),:]
```

```python
# slice(None)可以筛选出这一索引的所有内容
stocks.loc[(slice(None),["2023-03-16","2023-03-17"]),:]
```

```python
# 可以重置索引，把它变成普通的列
stocks.reset_index()
```
